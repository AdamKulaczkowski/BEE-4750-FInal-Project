```julia
function gac(Co,x,v,t,D,f_gac)
        k_d = k_gac*f_gac
        p_d = 1.3
        theta = 0.5
        R = 1 + (k_d*p_d)/theta 
        C = (Co)*erfc((x -(v/R)*t)/(2*sqrt((D/R)*t)))
    return C
end
```
