```julia
import Pkg; Pkg.add("SpecialFunctions"); 
using SpecialFunctions
using Plots; plotly();
```

```julia
function conc(Co,x,v,t,D)
    C = (Co/2)*erfc((x - v*t)/(2*sqrt(D*t)))
    return C
end
```

```julia
function ConcCurve(Co,x,v,D)
    j = zeros(9001)
    k = 0:9000
    for t in 1:9000
        i = conc(Co,x,v,t,D)
        j[t+1] = i[1]
    end
    return j,k
end
```

```julia
trial1 = ConcCurve(2878.5,0.5,0.0000000317,0.0001268392)
```
